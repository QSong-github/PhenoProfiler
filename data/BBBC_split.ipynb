```python
df_BBBC_filtered = df_BBBC.groupby(['Source', 'Metadata_Plate','Metadata_Well']).size().reset_index(name='Count')
df_BBBC_filtered.shape
```

```python
df_BBBC_filtered_specific = df_BBBC[
    (df_BBBC['Source'] == 'bbbc036') &
    (df_BBBC['Metadata_Plate'] == 26247)
]
df_BBBC_filtered_specific
```

```python
df_BBBC['Train'] = df_BBBC.groupby(['Source', 'Metadata_Plate', 'Metadata_Well'])['Metadata_Site'].cumcount().gt(0).astype(int)
```

```python
df_BBBC['Train'].value_counts()
```

```python
# 保存 Train == 0 的行到新的 CSV 文件
df_BBBC[df_BBBC['Train'] == 0].to_csv('/data/boom/bbbc_test.csv', index=False)

# # 保存 Train == 1 的行到新的 CSV 文件
df_BBBC[df_BBBC['Train'] == 1].to_csv('/data/boom/bbbc_train.csv', index=False)
```
